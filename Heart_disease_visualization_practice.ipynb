## Histogram:

### Purpose
To understand the distribution of a single numerical variable.

```python
plt.figure(figsize=(10,6))
plt.hist(df["trestbps"], bins=25)
plt.title('Distribution of Resting Blood Pressure')
plt.xlabel('Blood Pressurer')
plt.ylabel('Frequency')
plt.show()
```

```python
df.groupby('target')['trestbps'].mean()
```

```python
df['trestbps'].skew()
```

### Insight
- The distribution is right-skewed.
- Most observations fall within the range of 110 mmHg and 140 mmHg.
- A small number of patience shows hivalues above 160 mmHg, suggesting outliers.

```python
plt.figure()
plt.hist(df['thalach'], bins=25)
plt.title('Distribution of Maximum Heart Rate Achieved')
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')

plt.show()
```

```python
df.groupby('target')['thalach'].mean()
```

```python
df['thalach'].skew()
```

### Insight
- The distribution is approximately symmetric, centered around 150-170 bps.
- It also shows slight left tail.

## Count Plot:
### Purpose
To show the frequency of categorical values(target & sex).

```python
sns.countplot(
    data=df,
    x='target',
    palette="colorblind"
).set_xticklabels(['No Disease', 'Disease'])
```

```python
sns.countplot(
    data=df,
    x='sex',
    palette='colorblind'
).set_xticklabels(['Female', 'Male'])
```

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x='target',
    palette='colorblind',
    hue='sex'
)

ax.set_xticklabels(['No Disease', 'Disease'])
ax.set_xlabel("Heart Disease Status")
ax.set_ylabel('Number of Patients')
ax.legend(
    ['Female', 'Male'],
    title='Sex'
    )

plt.show()
```

### Insight
- Class imbalance is observed as we have more male sample data

## Scatter Plot:

### Purpose
To examine the relationship between two numerical variables.

```python
plt.figure()

max_hr_by_age = sns.scatterplot(
    data=df,
    x='age',
    y='thalach',
    hue='target',
    #style='sex',
    palette='colorblind',
    alpha=0.7
    )

max_hr_by_age.set_ylabel('Maximum Heat Rate Achieved')
handles, _ = max_hr_by_age.get_legend_handles_labels()
max_hr_by_age.legend(
    handles,
    ['Not Disease', 'Disease'],
    title='Heart Disease'
    )
plt.show()
```

```python
plt.figure()

sns.regplot(
    data=df[df['target'] == 0],
    x='age',
    y='thalach',
    scatter_kws={'alpha':0.4},
    label='No Disease'
)

sns.regplot(
    data=df[df['target'] == 1],
    x='age',
    y='thalach',
    scatter_kws={'alpha': 0.4},
    label='Disease'
)

plt.ylabel('Maximum Heart Rate (Regression Line)')
plt.title('Age vs Max Heart Rate (Regression Line)')
plt.show()
```

### Insight
- A negative relationship is observed between age and maximum heart rate in both group.
- This scatter plot suggests a possible difference in maximum heart rate between patients with and without heart disease. However due to differences in sample size and potential other factors such as age and sex, this visualization alone is insufficient to conclude that heart disease patients achieve higher maximum heart rate.

## Box Plot:
### Purpose
To compare distributions and identify outliers across groups.

```python
plt.figure()

sns.boxplot(
    data=df,
    x='target',
    y='thalach',
    palette='colorblind'
)

plt.xticks([0, 1], ['No Disease', 'Disease'])
plt.xlabel('Hear Disease Status')
plt.ylabel('Max Heart Rate Achieved')
plt.title('Distribusion of Max Heart Rate by Heart Disease Status')
plt.show()
```

## Violin Plot:
### Purpose
To visualize the full distribution shape across categories.

```python
plt.figure()

sns.violinplot(
    data=df,
    x='target',
    y='thalach',
    palette='colorblind',
    inner='quartile'
)

plt.xticks([0, 1], ['No Disease', 'Disease'])
plt.xlabel('Heart Disease Status')
plt.ylabel('Max Heart Rate Achieved')
plt.title('Violin Plot of Max Heart Rate by Heart Disease Status')

plt.show()
```

### Insight
- Density is concentrated around 140 for non disease group and 160 for disease group.

## Heatmap:
### Purpose
To visualize correlation strength between numerical variables.

```python
numeric_cols = [
    'age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'target'
]
corr = df[numeric_cols].corr()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    center=0,
    square=True,
    linewidths=0.5
)

plt.title('Correlation Matrix of Key Clinical Variables')
plt.show()
```

### Insight
- The heatmap shows that maximum heart rate (thalach) has a moderate negative correlation with age, indicating that older patients tend to achieve lower maximum heart rates.


## Pair Plot:
### Purpose
To explore multi-variable relationships simultaneously.

```python
pair_cols = [
    'age', 'trestbps', 'thalach', 'oldpeak', 'target'
]
```

```python
sns.pairplot(
    df[pair_cols],
    hue='target',
    palette='colorblind',
    diag_kind='kde'
)
```

###Pair Plot  Insights

The pair plot shows that patients with and without heart disease are not cleanly separated by any single variable.
Overall, the two groups largely overlap.

However, some tendencies are visible:

- oldpeak: Patients with heart disease appear more frequently at moderate to higher oldpeak values, while non-disease cases peak near zero but show wider spread.

- thalach: Non-disease cases tend to have slightly higher central values, but the overlap is substantial, making this feature weak on its own.

- age: Both groups are widely mixed across ages.

## KDE Plot:
### Purpose
To visualize a smoothed distribution of numerical variables.


```python
plt.figure()

sns.kdeplot(
    data=df,
    x='thalach',
    hue='target',
    fill=True,
    common_norm=False,
    palette='colorblind',
    alpha=0.6
)

plt.xlabel('Maximum Heart Rate Achieved')
plt.ylabel('Density')
plt.ylabel('Density')
plt.title('Distribution of Maximum Heart Rate by Heart Disease Status')
plt.legend(title='Heart Disease', labels=['No Disease', 'Disease'])

plt.show()
```

### Insight
- The KDE plot shows that patients with heart disease tend to have their distribution shifted toward higher max heart rate compared to those without disease.
- Although the distrubutions overlap, the density peak for the disease group occurs at a higher range, suggesting systematic difference rather than random variation.

## Line Plot:

### Purpose
To observe how a numerical variable changes when the data is ordered by another variable.
This is not a true time-series.

```python
age_bins = range(20, 85, 5)

df['age_bin'] = pd.cut(
    df['age'],
    bins=age_bins
)

grouped = (
    df
    .groupby(['age_bin', 'target'])['thalach']
    .mean()
    .reset_index()
)
grouped['age_bin'] = grouped['age_bin'].astype(str)

plt.figure()

ax = sns.lineplot(
    data=grouped,
    x = 'age_bin',
    y = 'thalach',
    hue='target',
    hue_order=[0,1]
)

handles, _ = ax.get_legend_handles_labels()
ax.legend(
    handles,
    ['No Disease', 'Disease'],
    title='Heart Disease Status'
)

ax.set_title(
    'Averaged Max Heart Rate by Age Group'
)
ax.set_xlabel('Age Group')
ax.set_ylabel('Averaged Max Heart Rate')

plt.xticks(rotation=45)
plt.show()
```

### Insight
- For patients with heart disease, maximum heart rate generally decreases as age increases.

```python
op_grouped = (df
              .groupby(['age_bin', 'target'])['oldpeak']
              .mean()
              .reset_index()
              )

op_grouped['age_bin'] = op_grouped['age_bin'].astype(str)

plt.figure()

ax = sns.lineplot(
    data=op_grouped,
    x='age_bin',
    y='oldpeak',
    hue = 'target',
    hue_order=[0,1]
)

handles, _ = ax.get_legend_handles_labels()
ax.legend(
    handles,
    ['No Disease', 'Disease'],
    title='Heart Disease Status'
)

ax.set_title(
    'Averaged Oldpeak by Age Group'
)
ax.set_xlabel('Age Group')
ax.set_ylabel('Averaged Oldpeak')

plt.xticks(rotation=45)
plt.show()
```

### Insight

- Oldpeak shows a slight increase related to age in the disease group.
- the non-disease group shows more volatile values with sharp dicrease with patients at 60 or higher
